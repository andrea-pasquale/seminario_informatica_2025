### Liste (equivalenti agli array in C++)

````python
# creiamo un array vuoto
v = []
print(v)

# creiamo un array di cinque elementi
v = [10, 20, 30, 40, 50]
print("Lista:", v)
print("Primo elemento:", v[0])
print("Lunghezza:", len(v))

# Iterazione su una lista
for i in range(len(v)):  # parti da 0 e arriva a len(v) - 1, incrementa di 1 -> for(int i = 0; i < len; i++)
    print("Elemento i-esimo:", v[i])

for i in range(len(v)):
    print("Elemento numero", i, ":", v[i])

# Aggiunta elementi
v.append(60)
print("v con un elemento aggiunto alla fine:", v)

# Modifica elementi
v[0] = 5
print(v)
````

### Condizioni if/else/else if

Versione in C++:
```c++
include <iostream>
include <cmath>
using namespace std;

int main()
{
  // dichiarazioni
  double a, b, c;

  // assign values
  cout << "Introdurre i coefficienti a b c (separati da spazio):\n";
  cin >> a >> b >> c;

  cout << "Soluzione per " << a << "*x^2 + " << b << "*x + " << c << " = 0.\n";
  
  // calcolare discriminante
  const double D = pow(b, 2) - 4 * a * c;

  if (D > 0)
    {
      const double x1 = (-b + sqrt(D)) / (2 * a);
      const double x2 = (-b - sqrt(D)) / (2 * a);
      cout << "Soluzione x1 = " << x1 << endl;
      cout << "Soluzione x2 = " << x2 << endl;      
    }
  else if (D == 0)
    {
      const double x12 = -b / (2 * a);
      cout << "Soluzione x1,2 = " << x12 << endl;
    }
  else // D < 0
    {
      const double real = -b / (2 * a);
      const double imag1 = sqrt(-D) / (2 * a);
      const double imag2 = -imag1;
      cout << "Soluzione x1 = " << real << " + i * " << imag1 << endl;
      cout << "Soluzione x2 = " << real << " + i * " << imag2 << endl;
    }
    
  return 0;
}
```

Verifichiamo il risultato per i seguenti coefficienti:
  - $D > 0$: $a = 2, b = 5, c = 2 \to x_1 = -0.5, x_2 = -2.0$.
  - $D = 0$: $a = 4, b = -4, c = 1 \to x_{1,2} = 0.5$.
  - $D < 0$: $a = 1, b = 4, c = 5 \to x_1 = -2 + i, x_2 = -2 - i$.

````python
import numpy as np

a = 2
b = 5
c = 2
# a = 4
# b = -4
# c = 1
# a = 1
# b = 4
# c = 5

print("Soluzione per", a, "* x^2 +", b, "* x +", c)

discriminante = b**2 - 4 * a * c

if discriminante > 0:
    x1 = (-b + np.sqrt(discriminante)) / (2 * a)
    x2 = (-b - np.sqrt(discriminante)) / (2 * a)
    print("Soluzione x1:", x1)
    print("Soluzione x2:", x2)
elif discriminante == 0:
    x12 = -b / (2 * a)
    print("Soluzione x12:", x12)
else:
    print("Nessuna soluzione reale")
    reale = -b / (2 * a)
    immaginaria1 = np.sqrt(-discriminante) / (2 * a)
    immaginaria2 = -immaginaria1
    print("Soluzione x1:", reale, "+ i *", immaginaria1)
    print("Soluzione x2:", reale, "+ i *", immaginaria2)
````

### Cicli for/while

Versione C++:
```c++
#include <iostream>
using namespace std;

int main()
{
  // ciclo for
  for (int i = 0; i < 5; i++)
    {
      cout << "Hello World for index = " << i << endl;
    }
  // ciclo while
  int counter = 0;
  while(counter < 5)
    {
      cout << "Hello World while index = " << counter << endl;
      counter++;
    }
  return 0;
}
```

Versione Python

````python
for i in range(5):
    print("Hello World for index =", i)
````

````python
counter = 0
while counter < 5:
    print("Hello World while index =", counter)
    counter += 1  # in python non esiste l'operatore ++
````

Versione C++:
```c++
#include <iostream>
#include <fstream>
#include <cmath>
using namespace std;

#define N 1000

int main()
{
  // creazione variabile per la velocita;
  double v[N];
  double spazio, tempo;

  // apertura file dati
  fstream f;
  f.open("../data_moto.dat", ios::in);

  if (!f.good())
    {
      cerr << "Errore lettura file data_moto.dat" << endl;
      return 1;
    }

  // lettura dati
  for (int i = 0; i < N; i++)
    {
      f >> spazio >> tempo;
      v[i] = spazio / tempo;
    }
  
  f.close();

  // Calcolo velocità media
  double sum = 0;
  for (int i = 0; i < N; i++)
    sum += v[i];

  double vmedia = sum / N;
  cout << "Velocità media = " << vmedia << endl;

  // Calcolo dev.std di v
  double sum2 = 0;
  for (int i = 0; i < N; i++)
    sum2 += pow(v[i] - vmedia, 2);

  double sigma = sqrt(sum2/(N-1));
  cout << "Dev. standard = " << sigma << endl;

  // calcolo la velocità minima
  double vmin = v[0];
  for (int i = 1; i < N; i++)
    if (v[i] < vmin)
      vmin = v[i];

  cout << "Velocità minima = " << vmin << endl;

  // calcolo la velocità massima
  double vmax = v[0];
  for (int i = 1; i < N; i++)
    if (v[i] > vmax)
      vmax = v[i];

  cout << "Velocità massima = " << vmax << endl;
  
  return 0;
}
```

Versione Python

````python
import numpy as np

f = open("data_moto.dat")

num_righe = 1000
velocita = np.zeros(num_righe)
for i in range(num_righe):
    line = f.readline()
    split_line = line.split(" ")
    spazio = float(split_line[0])
    tempo = float(split_line[1])
    velocita[i] = spazio / tempo
f.close()

print(velocita)
````

````python
media = 0
for i in range(len(velocita)):
    media += velocita[i]
media /= len(velocita)
print("Velocità media:", media)

sigma = 0
for i in range(len(velocita)):
    sigma += velocita[i]**2 - media
sigma = np.sqrt(sigma / (len(velocita) - 1))
print("Dev. standard = ", np.std(velocita))

vmin = velocita[0]
for i in range(1, len(velocita)):
    if velocita[i] < vmin:
        vmin = velocita[i]
print("Velocità minima =", vmin)

vmax = velocita[0]
for i in range(1, len(velocita)):
    if velocita[i] > vmax:
        vmax = velocita[i]
print("Velocità massima =", vmax)
````

````python
# in alternativa
print("Velocità media:", np.mean(velocita))
print("Dev. standard = ", np.std(velocita))
print("Velocità minima =", np.min(velocita))
print("Velocità massima =", np.max(velocita))
````

````python
import matplotlib.pyplot as plt
plt.plot(velocita)
plt.hlines(np.mean(velocita), 0, len(velocita), color="red", label="media")
plt.hlines([np.mean(velocita) + np.std(velocita), np.mean(velocita) - np.std(velocita)], 0, len(velocita), color="red", linestyles="--", label="dev.std.")
plt.hlines(np.max(velocita), 0, len(velocita), color="green", label="massimo")
plt.hlines(np.min(velocita), 0, len(velocita), color="purple", label="minimo")
plt.xlabel("data point")
plt.ylabel("km/h")
plt.legend()
plt.show()
````
